```python
import pandas as pd

# Step 1: Read the CSV file into a DataFrame
cols = ['text1', 'text2']
path = "C:/Users/shric/Desktop/Test Semi/Question.csv"
df = pd.read_csv(path, names=cols, skiprows=1)

# Step 2: Get user input
user_ques_1 = input("Enter text to add to the CSV file: ")
user_ques_2 = input("Enter text to add to the CSV file: ")


# Step 3: Append the user input to the DataFrame
new_row = {'text1': user_ques_1, 'text2': user_ques_2}
df = df.append(new_row, ignore_index=True)

# Step 4: Write the updated DataFrame back to the CSV file
df.to_csv(path, index=False, header=False, mode='a')

print("User input has been added to the CSV file.")

# Display the updated DataFrame
print(df)


```
